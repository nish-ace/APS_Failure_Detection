```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
import pickle
```

```python
df = pd.read_csv(r'E:\iNeuron\Self Project\prediction_file_from_DB\aps_test_set.csv')
df
```

```python
df.replace('na', np.nan, inplace=True)
```

```python
X,y = df.drop(columns = ['class']), df['class']
```

```python
for _ in X.columns:
    X[_] = X[_].astype('float64')
```

```python
imp = SimpleImputer(strategy = 'median')
X = pd.DataFrame(imp.fit_transform(X), columns=X.columns)
X
```

```python
X.drop(columns=['Idx'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_y, test_y = train_test_split(X,y,random_state=45,test_size=0.3)
```
