### Training Data

```python
df_train = pd.read_csv(r'E:\iNeuron\APSFailureDetection\training_file_from_DB\aps_training_set.csv')
df_train
```

```python
smote = SMOTE()
```

```python
df_train.replace('na', np.nan, inplace=True)
```

```python
X,y = df_train.drop(columns = ['class']), df_train['class']
```

```python
for _ in X.columns:
    X[_] = X[_].astype('float64')
```

```python
imp = SimpleImputer(strategy = 'median')
X = pd.DataFrame(imp.fit_transform(X), columns=X.columns)
X
```

```python
X.drop(columns=['Idx'], inplace=True)
```

```python
X_,y_ = smote.fit_resample(X,y) 
```

```python
y_.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_y, test_y = train_test_split(X_,y_,test_size = 0.3, random_state = 50)
```

### Extra

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import plot_confusion_matrix
```

```python
param_grid = {'n_estimators':800,
              'max_depth':13,
              'min_samples_split':12,
              'max_leaf_nodes':300,
              'ccp_alpha':0.01,
              'criterion': 'gini'}

rf = RandomForestClassifier(**param_grid)
rf.fit(train_X,train_y)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
def convert(arr):
    ls = []
    for _ in arr:
        if _ == False:
            ls.append(1)
        else:
            ls.append(-1)
    return np.array(ls)
```

```python
predicted_proba = rf.predict_proba(test_X)
y_predicted = (predicted_proba [:,1] >= 0.0196)
y_predicted = convert(y_predicted)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_true=test_y, y_pred=y_predicted).ravel()
```
